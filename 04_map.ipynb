# 04 Repeated Choreopleth Map

Let's look at our final question:
## Q4: How did the trend spread across Europe?

The idea is to create repeated choropleth maps to reflect the trend spreading across Europe. Soo has kindly pointed me to some choropleth map examples by altair:
- [US Income by State: Wrapped Facet](https://altair-viz.github.io/gallery/us_incomebrackets_by_state_facet.html)
- [Repeated Choropleth Map](https://altair-viz.github.io/gallery/choropleth_repeat.html)

But actually, I found out that [DataWrapper](https://app.datawrapper.de/) has more beautiful and easy-to-use data viz options! So instead, we're going to use pandas to export the csv file that we need, and then make data visualization in DataWrapper.

### import modules and dataset

```python
import pandas as pd
```

```python
trials_clean = pd.read_csv('./export/trials_clean.csv')
trials_clean
```

```python
# Subsetting the dataframe for datawrapper. We only need century, tried, and country
trials_simplified = trials_clean[['century', 'tried', 'gadm.adm0']]
trials_simplified.sample(20)
```

```python
# export by century
trials_simplified_13 = trials_simplified.loc[trials_simplified['century']==1300]
trials_simplified_14 = trials_simplified.loc[trials_simplified['century']==1400]
trials_simplified_15 = trials_simplified.loc[trials_simplified['century']==1500]
trials_simplified_16 = trials_simplified.loc[trials_simplified['century']==1600]
trials_simplified_17 = trials_simplified.loc[trials_simplified['century']==1700]
trials_simplified_18 = trials_simplified.loc[trials_simplified['century']==1800]
```

```python
trials_simplified_13.to_csv('./export/trials_simplified_13.csv', index=False)
trials_simplified_14.to_csv('./export/trials_simplified_14.csv', index=False)
trials_simplified_15.to_csv('./export/trials_simplified_15.csv', index=False)
trials_simplified_16.to_csv('./export/trials_simplified_16.csv', index=False)
trials_simplified_17.to_csv('./export/trials_simplified_17.csv', index=False)
trials_simplified_18.to_csv('./export/trials_simplified_18.csv', index=False)
```
